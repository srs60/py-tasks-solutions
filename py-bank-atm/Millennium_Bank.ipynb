```python
# demo data
cAccs = ['12345678','24680246','35791357','12378901','36789122']
cNames = ['Smith', 'John', 'Tom', 'Anna', 'James']
cPins = ['1234', '1367', '5678', '3612', '3579']
cBalances = [10000, 20000, 25000, 40000, 50000]
cInterestRates = [1.5,1.5,1.5,1.6,1.7]
data = list(zip(cAccs,cNames,cPins,cBalances,cInterestRates))
#for d in data: print(d)
```

```python
class Bank_Account(object): 
    
    AdminStatus = False
    
    def __init__(self):
        self.accNum = ""
        self.accHolder = ""
        self.accPin = ""
        self.accType = "D"    # Deposit/Credit
        self.accBalance = 0
        self.accIRate = 0
        self.Open = False
        self.clientNum = -1
        
    def openAccount(self, n):
        self.clientNum = n
        self.accNum = data[n][0]
        self.accHolder = data[n][1]
        self.accPin = data[n][2]
        self.accType = "D"           # Deposit/Credit
        self.accBalance = data[n][3]
        self.accIRate = data[n][4]
        
    def changePin(self, old_pin, new_pin1, new_pin2):
        if old_pin == self.accPin:
            if new_pin1 == new_pin2:
                self.accPin = new_pin1
                clientData = (self.accNum, self.accHolder, self.accPin, self.accBalance, self.accIRate)
                data[self.clientNum] = clientData
                #for d in data: print(d)
        
    def deposit(self, amount):
        self.accBalance += amount 
        print("\n Amount Deposited:",amount) 

    def withdraw(self, amount):
        if self.accBalance >= amount: 
            self.accBalance -= amount 
            print("\n Your Withdrew:", amount)
            print(" Your Balance: ", self.accBalance)            
        else: 
            print("\n Insufficient Balance ") 

    def report(self):
        #print(self.accNo, " ",self.name ," ",self.type," ", self.deposit)  
        print(f'{self.accNum} {self.accHolder} {self.type} {self.accBalance}')  

    def display(self): 
        print("\n Available Balance:",self.accBalance) 
```
